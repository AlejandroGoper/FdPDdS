# Programa para el uso y manejo de la FFT 

Calcularemos la FFT de una señal cosenoidal con frecuencia 2 Hz

```python
%matplotlib inline
from numpy.fft import fft
from numpy import arange,pi,cos,abs,angle
import matplotlib.pyplot as plt 
```

Primero llenamos una secuencia con los valores de t en los que queremos evaluar cos(t);

DeltaT se refiere al tiempo que transcurre entre medicion y medicion; Consideremos que esta en segundos

```python
DeltaT = 0.1    # Esto se refiere al periodo de muestre en el dominio del tiempo
tiempo_total = 2    # Este es el tiempo total de medición
t = arange(0,tiempo_total,DeltaT)   # Análogo a 0:DeltaT:tiempo_total-deltaT de MATLAB
```

f es la frecuencia de la señal discretizada en ciclos / unidad de tiempo --> Hz
w que es la frecuencia angular, se relaciona con f como:

w = 2 * PI * f

```python
f = 2   #Esta es la frecuencia de la señal cosenoidal (recordemos que queremos 2 Hz)
w = 2*pi*f  # Frecuencia angular (de una señal cosenoidal -- cos(2pift))
```

### Construyendo señal cosenoidal

```python
f1 = cos(w*t)   
```

### Graficando valores en el dominio del tiempo

```python
plt.plot(t,f1)  # en forma de lineas continuas
```

```python
plt.stem(t,f1) # en forma de secuencia
```

### Calculando la FFT de la señal

```python
F = fft(f1) 
```

### Graficando la FFT de la señal 

Recordemos que las unidades del eje X son ciclos/tiempo_copleto_medicion

En este caso dado que tiempo_total = 2s -- las unidades son 0.5 Hz


```python
plt.stem(abs(F),use_line_collection=True)    #Graficamos la magnitud 
plt.title("Magnitud |F|")
plt.xlabel("ciclos/tiempo_total")
```

```python
plt.stem(angle(F)) # Graficamos la fase 
plt.title("Fase arg(F)")
```

## Dandole unidades a la grafica de la FFT

Para darle unidades de potencia a la magnitud de F, es necesario, dividir la contribución de cada punto del arreglo entre el numero total de datos (dado que la FFT por defaut, es tal que la suma de todas las contribuciones del arreglo es igual a la logitud del arreglo), de esta manera, lograremos normalizar.

```python
plt.stem(abs(F)/len(F),use_line_collection=True)
plt.title("Magnitud")
plt.xlabel("Ciclos/tiempo_total")
plt.ylabel("Potencia")
```
